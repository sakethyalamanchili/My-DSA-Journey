# Dictionaries

Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

```python
# Creating Dictionaries
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict=dict()
empty_dict
```

```python
student = {"name": "John", "age": 21, "courses": ["Math", "CompSci"]}
print(student)
print(student["name"])
print(student["age"])
print(student["courses"])
print(student["courses"][0])
print(student["courses"][1])
print(len(student))
print(student.keys())
print(student.values())
print(student.items())

print(type(student))
```

```python
# Using get() method, to avoid KeyError if key not found
print(student.get('name'))
print(student.get('lastname'))  # returns None
print(student.get('lastname', 'Not Found'))  # returns 'Not Found'
```

```python
# Modifying Dictionaries
student['age'] = 22
student['address'] = "USA"
print(student)
```

```python
del student['address']
print(student)
```

```python
# Dictionary Methods

keys = student.keys()
print(keys)
values = student.values()
print(values)

items = student.items()
print(items)
```

```python
# Shallow Copying
student2 = student.copy()
print(student2)
print(student)
```

```python
student['name'] = "Saketh"
print(student)
print(student2)
```

```python
# Iterating over Keys
for key in student.keys():
    print(key)
```

```python
# Iterating over Values

for value in student.values():
    print(value)
```

```python
# Iterating over Key-Value Pairs

for key, value in student.items():
    print(f"Key: {key}, Value: {value}")
```

```python
# Nested Dictionaries
students = {
    "student1": {"name": "John", "age": 21},
    "student2": {"name": "Jane", "age": 22},
    "student3": {"name": "Doe", "age": 20}
}
print(students)
```

```python
# Accessing nested dictionaries elements 
print(students["student1"]["name"])  # Output: John
print(students["student2"]["age"])   # Output: 22
```

```python
# Iterating over nested dictionaries
for student_id, student_info in students.items():
    print(f"{student_id}: {student_info}")
    for key, value in student_info.items():
        print(f"{key}: {value}")
```

```python
# Dictionary Comphrehension
squares = {x : x ** 2 for x in range(6)}
print(squares) 
```

```python
# Condition dictionary comprehension
even_squares = {x : x ** 2 for x in range(6) if x % 2 == 0}
print(even_squares)
```

# Practical Examples

## Use a dictionary to count the frequency of elements in list

```python
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1

print(frequency)
```

## Merge 2 Dictionaries into one

```python
dict_1 = {'a': 1, 'b': 2}
dict_2 = {'b': 3, 'c': 4}

merged_dict = {**dict_1, **dict_2}
print(merged_dict)  # Output: {'a': 1, 'b': 3, 'c': 4}
```
