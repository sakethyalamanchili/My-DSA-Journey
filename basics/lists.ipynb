```python
# Accessing List Elements
fruits = ["Apple", "Banana", "Cherry", "Kiwi", "Blueberry"]
```

```python
print(fruits[0])     # First element
print(fruits[2])     # Third element
print(fruits[-1])    # Last element
```

```python
print(fruits[0:2]) # First two elements
print(fruits[-1:2:-1]) # Last to third element in reverse order
```

```python
fruits[1] = "Blackcurrant"
print(fruits)
```

```python
fruits.append("Watermelon")
print(fruits)
```

```python
fruits.insert(2, "Orange")
print(fruits)
```

```python
fruits.remove("Kiwi")
print(fruits)
```

```python
popped_fruit = fruits.pop()
print(popped_fruit)
print(fruits)
```

```python
index = fruits.index("Cherry")
print(index)
```

```python
fruits.insert(2, 'Mango')
print(fruits.count('Mango'))
```

```python
# Iterating Over List

for number in range(1, 11):
    print(number)
```

```python
# Iterating with Index

for index, number in enumerate(range(1, 11)):
    print(f"Index: {index}, Number: {number}")
```

```python
# List Comprehensions

lst = []
for x in range(1, 11):
    lst.append(x**2)
    
print(lst)
```

```python
[x**2 for x in range(1, 11)]
```

```python
# Nested List Comprehension

lst_1 = [1, 2, 3, 4, 5]
lst_2 = ["one", "two", "three", "four", "five"]

pair = [[i,j] for i in lst_1 for j in lst_2]
print(pair)
```

```python
# List Comprehension with function calls

words = ["hello", "world", "python", "programming"]
lengths = [len(word) for word in words]
print(lengths)
```
